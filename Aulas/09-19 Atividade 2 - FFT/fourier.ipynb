```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

mpl.rcParams['axes.titlesize'] = 10
mpl.rcParams['axes.labelsize'] = 8
mpl.rcParams['xtick.labelsize'] = 8
mpl.rcParams['ytick.labelsize'] = 8
```

```python
f = 3

tt = np.linspace(0, 10, 1000, True)
tf = np.linspace(0, 10, 1000, False)

ft = np.sin(2*np.pi*f*tt)
ff = np.sin(2*np.pi*f*tf)

ftt = np.fft.fft(ft)
ftf = np.fft.fft(ff)

dtt = np.array([ft @ np.e**(-1j*2*np.pi*i*np.arange(len(tt))/len(tt)) for i in range(len(tt))])/len(tt)
dtf = np.array([ff @ np.e**(-1j*2*np.pi*i*np.arange(len(tf))/len(tf)) for i in range(len(tf))])/len(tf)
```

```python
fig, axis = plt.subplots(2, 3, figsize=(18, 12))
fig.set_tight_layout(True)

axis[0, 0].plot(tt, ft)
axis[0, 1].plot(tt, np.abs(ftt))
axis[0, 2].plot(tt, np.abs(dtt))

axis[1, 0].plot(tf, ff)
axis[1, 1].plot(tf, np.abs(ftf))
axis[1, 2].plot(tf, np.abs(dtf))
```
