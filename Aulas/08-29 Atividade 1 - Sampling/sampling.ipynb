# Tarefa em Sala

* Dado o sinal analógico $f(x,y) = 256\sin(2\pi(50x+70y))$
* Quais são as frequências utilizadas?
* Representar na forma de Euler.
* Visualize o sinal usando scipy (matplotlib)
* Agora faça a amostragem com:
  * $60\times60$, $100\times100$, $300\times300$ e $1000\times1000$ amostras
  * Mostre o resultado com imagens em tons de cinza
* Quantas amostras são necessárias para reconstruir o sinal analógico?

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

from mpl_toolkits.mplot3d import Axes3D

mpl.rcParams['axes.titlesize'] = 10
mpl.rcParams['axes.labelsize'] = 8
mpl.rcParams['xtick.labelsize'] = 8
mpl.rcParams['ytick.labelsize'] = 8
```

```python
def f(x, y):
    return 256*np.sin(2*np.pi*(50*x+70*y))
```

```python
fig, axis = plt.subplots(1, 4, figsize=(20, 6))
fig.set_tight_layout(True)

x, y = np.mgrid[0:1:60j, 0:1:60j]
axis[0].pcolormesh(f(x, y))

x, y = np.mgrid[0:1:100j, 0:1:100j]
axis[1].pcolormesh(f(x, y))

x, y = np.mgrid[0:1:300j, 0:1:300j]
axis[2].pcolormesh(f(x, y))

x, y = np.mgrid[0:1:1000j, 0:1:1000j]
axis[3].pcolormesh(f(x, y))
```

```python
fig, axis = plt.subplots(2, 2, figsize=(20, 12))
fig.set_tight_layout(True)

x, y = np.mgrid[0:1:60j, 0:1:60j]
axis[0, 0].remove()
axis[0, 0] = fig.add_subplot(2, 2, 1, projection='3d')
axis[0, 0].scatter(x, y, f(x, y))

x, y = np.mgrid[0:1:100j, 0:1:100j]
axis[0, 1].remove()
axis[0, 1] = fig.add_subplot(2, 2, 2, projection='3d')
axis[0, 1].scatter(x, y, f(x, y))

x, y = np.mgrid[0:1:300j, 0:1:300j]
axis[1, 0].remove()
axis[1, 0] = fig.add_subplot(2, 2, 3, projection='3d')
axis[1, 0].scatter(x, y, f(x, y))

x, y = np.mgrid[0:1:1000j, 0:1:1000j]
axis[1, 1].remove()
axis[1, 1] = fig.add_subplot(2, 2, 4, projection='3d')
axis[1, 1].scatter(x, y, f(x, y))
```

```python
fig, axis = plt.subplots(2, 2, figsize=(20, 12))
fig.set_tight_layout(True)

x, y = np.mgrid[0:1:60j, 0:1:60j]
axis[0, 0].remove()
axis[0, 0] = fig.add_subplot(2, 2, 1, projection='3d')
axis[0, 0].plot_wireframe(x, y, f(x, y))

x, y = np.mgrid[0:1:100j, 0:1:100j]
axis[0, 1].remove()
axis[0, 1] = fig.add_subplot(2, 2, 2, projection='3d')
axis[0, 1].plot_wireframe(x, y, f(x, y))

x, y = np.mgrid[0:1:300j, 0:1:300j]
axis[1, 0].remove()
axis[1, 0] = fig.add_subplot(2, 2, 3, projection='3d')
axis[1, 0].plot_wireframe(x, y, f(x, y))

x, y = np.mgrid[0:1:1000j, 0:1:1000j]
axis[1, 1].remove()
axis[1, 1] = fig.add_subplot(2, 2, 4, projection='3d')
axis[1, 1].plot_wireframe(x, y, f(x, y))
```

```python
fig, axis = plt.subplots(2, 2, figsize=(20, 12))
fig.set_tight_layout(True)

x, y = np.mgrid[0:1:60j, 0:1:60j]
axis[0, 0].remove()
axis[0, 0] = fig.add_subplot(2, 2, 1, projection='3d')
axis[0, 0].plot_surface(x, y, f(x, y))

x, y = np.mgrid[0:1:100j, 0:1:100j]
axis[0, 1].remove()
axis[0, 1] = fig.add_subplot(2, 2, 2, projection='3d')
axis[0, 1].plot_surface(x, y, f(x, y))

x, y = np.mgrid[0:1:300j, 0:1:300j]
axis[1, 0].remove()
axis[1, 0] = fig.add_subplot(2, 2, 3, projection='3d')
axis[1, 0].plot_surface(x, y, f(x, y))

x, y = np.mgrid[0:1:1000j, 0:1:1000j]
axis[1, 1].remove()
axis[1, 1] = fig.add_subplot(2, 2, 4, projection='3d')
axis[1, 1].plot_surface(x, y, f(x, y))
```
