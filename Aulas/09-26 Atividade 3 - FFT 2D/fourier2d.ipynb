## Aprender a carregar "Lena" em um array 2D numpy

```python
import matplotlib as mpl
import matplotlib.cm
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np

from matplotlib import colorbar
from matplotlib.colors import LogNorm, Normalize, to_rgb

from mpl_toolkits.axes_grid1 import make_axes_locatable

from PIL import Image

mpl.rcParams['axes.titlesize'] = 10
mpl.rcParams['axes.labelsize'] = 8
mpl.rcParams['xtick.labelsize'] = 8
mpl.rcParams['ytick.labelsize'] = 8
```

```python
def add_colorbar(ax, pos='right', size=0.1, pad=0.05, 
                 cmap=None, norm=None, off=False, 
                 orientation='vertical', sharex=None):
    divider = make_axes_locatable(ax)
    bar = divider.append_axes(pos, size, pad=pad, sharex=sharex)
    
    if isinstance(cmap, str):
        cmap = matplotlib.cm.cmap_d[cmap]

    if off:
        bar.axis('off')
    else:
        colorbar.ColorbarBase(bar, cmap=cmap, norm=norm, 
                              orientation=orientation)

    return bar
```

```python
word = 2**8-1
norml8 = Normalize(0, word)

img = np.asarray(Image.open('Lena.png').convert('L'), dtype=float)
print(img/word)

assert img.shape == (512, 512)

W, H = img.shape
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
fig.set_tight_layout(True)

def drawimg(ax, title, img, cmap, norm):
    ax.set_title(title)
    ax.imshow(img, cmap=cmap, norm=norm)
    add_colorbar(ax, cmap=cmap, norm=norm)
    
drawimg(ax, 'Original Image', img, 'gray', norml8)
```

## Implementar em Python a transformada DFT em 2D

```python
def dft2(arr):
    M, N = arr.shape
    m, n = np.mgrid[0:M, 0:N]
    
    arr_dft = np.full(arr.shape, 0)
    
    for u in range(0, M):
        for v in range(0, N):
            arr_dft[u, v] = np.sum(arr*np.exp(-2j*np.pi*(u*m/M+v*n/N)))
    
    return np.roll(np.roll(arr_dft, M//2, 0), N//2, 1)

img_dft = dft2(img[0:512:2, 0:512:2])

print(img_dft.shape)
print(img_dft)
```

## Comparar resultados com a rotina `numpy.fft.fft2()`

```python
normlgs = LogNorm(1, np.sum(img))
normlp = Normalize(0, np.log1p(np.sum(img)))

img_fft = np.fft.fftshift(np.fft.fft2(img))

assert np.max(np.log1p(np.abs(img_fft))) <= np.log1p(np.sum(img))

print(img_fft)
```

## Transformar "Lena" para o espaço de frequências

```python
fig, axis = plt.subplots(1, 3, figsize=(18, 6))
fig.set_tight_layout(True)

drawimg(axis[0], 'Original Image', img, 'gray', norml8)
drawimg(axis[1], 'SDFT2 Image', np.log1p(np.abs(img_dft)), 'viridis', normlp)
drawimg(axis[2], 'SFFT2 Image', np.log1p(np.abs(img_fft)), 'viridis', normlp)
```

## Filtrar com uma função Gaussiana com média 0 e desvios-padrão $\sigma_x$, $\sigma_y$

```python
norml0 = Normalize(0, 1)

def gauss(arr, sx, sy):
    X, Y = arr.shape
    x, y = np.mgrid[0:X, 0:Y]
    
    return np.exp(-((x-X/2)**2/(2*sx**2) + (y-Y/2)**2/(2*sy**2))**2)

flt_gss = gauss(img, np.sqrt(img.shape[0]), np.sqrt(img.shape[1]))
img_gss = img_fft*flt_gss

print(img_gss)
```

```python
fig, axis = plt.subplots(2, 3, figsize=(18, 12))
fig.set_tight_layout(True)

drawimg(axis[0, 0], 'Original Image', img, 'gray', norml8)
drawimg(axis[0, 1], 'SDFT2 Image', np.log1p(np.abs(img_dft)), 'viridis', normlp)
drawimg(axis[0, 2], 'SFFT2 Image', np.log1p(np.abs(img_fft)), 'viridis', normlp)
drawimg(axis[1, 0], 'Gaussian filter', flt_gss, 'gray', norml0)
drawimg(axis[1, 1], 'SFFT2 filtered Image', np.log1p(np.abs(img_gss)), 'viridis', normlp)
```

## Fazer a transformada inversa (frequência para imagem)

```python
img_ifft = np.real(np.fft.ifft2(np.fft.ifftshift(img_gss)))

print(img_ifft)
```

```python
fig, axis = plt.subplots(2, 3, figsize=(18, 12))
fig.set_tight_layout(True)

drawimg(axis[0, 0], 'Original Image', img, 'gray', norml8)
drawimg(axis[0, 1], 'SDFT2 Image', np.log1p(np.abs(img_dft)), 'viridis', normlp)
drawimg(axis[0, 2], 'SFFT2 Image', np.log1p(np.abs(img_fft)), 'viridis', normlp)
drawimg(axis[1, 0], 'Gaussian filter', flt_gss, 'gray', norml0)
drawimg(axis[1, 1], 'SFFT2 filtered Image', np.log1p(np.abs(img_gss)), 'viridis', normlp)
drawimg(axis[1, 2], 'Inverted SFFT2 filtered Image', img_ifft, 'gray', norml8)
```
